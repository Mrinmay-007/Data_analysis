```python
# find the top 10 batsman in terms of runs
IPL.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no. of sixes
six = IPL[IPL['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(3)
```

```python
# find the bastsman with most number of 4 and 6 in last 5 over 
ov = IPL[IPL['over']>15]
# type(ov)
ov = ov[(ov['batsman_runs']==6)|(ov['batsman_runs']==4)]
ov.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(3)
```

```python
# find kohli's record against all teams
# IPL.groupby('batsman')['batsman']
vk= IPL[IPL['batsman']=='V Kohli']
vk.groupby('bowling_team')['batsman_runs'].sum()
```

```python
# create a function that can return the heighest  score of any batsman
vk.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1)
```

```python
def highest(batsman):
    t_d = IPL[IPL['batsman']== batsman]
    return t_d.groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]
highest('MS Dhoni')
```
