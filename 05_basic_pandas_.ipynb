```python
# write a functiion  that can return the  track record of 2 teams  against each other

def func(t1,t2):
    df=ipl[((ipl['Team1']==t1) & (ipl['Team2']==t2)) |  ((ipl['Team1']==t2) & (ipl['Team2']==t2))]
    t1_win= df[df['WinningTeam']==t1].shape[0]
    t2_win=df[df['WinningTeam']==t2].shape[0]
    print(t1 + ' win: ' + str(t1_win))
    print(t2 + ' win: ' + str(t2_win))

func('Mumbai Indians','Kolkata Knight Riders')
```
